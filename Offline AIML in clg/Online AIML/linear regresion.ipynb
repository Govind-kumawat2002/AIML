```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from  sklearn.linear_model import LinearRegression 
#it is used for model (model bnane ke liye) 
                                                    #and linearregreesion is function which i used for create a model
from sklearn.preprocessing import PolynomialFeatures # 
from  sklearn.metrics import mean_absolute_error
from  sklearn.metrics import mean_squared_error  # check krta hai model shi hai ya glt
```

```python
np.random.seed(0) # random value find krne ke liye 
x = 2-3*np.random.normal(0,1,20) 
y = x-2*(x**2) + 0.5*(x**3)+ np.random.normal(-3,3,20) # it  is a equestion which is used for linear regreaaione
```

```python
plt.scatter(x,y) # x and y ke roop me graph bnane ke liye
```

```python
x= x.reshape(-1,1) #kitne bi kitnr ka array
y = y.reshape(-1,1)
model = LinearRegression() #class bnaiiii linear regression ke name se
model.fit(x,y) # present krne ke liye
y_pred = model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color = 'r')
plt.show
```

# New Data

```python
n = 1000
x1 = np.random.randint(1,10,n) #
x2 = np.linspace(0, .5, n)

noise = np.random.normal(0,0.5,n)
y = np.sin(x1) + np.cos(x2) + noise #sin eqestions 

df = pd.DataFrame({'x1':x1, 'x2':x2, 'y': y }) # data set 
df


```

```python
#lr modal 
model = LinearRegression()
model.fit(df[['x1','x2']],df['y'])
```

```python

x = df[['x1','x2']]
df['y_pred']= model.predict(x)
```

```python
! pip install plotly 
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(data=[go.Scatter3d(x= df['x1'], y = df['x2'], z = df['y'], mode = "markers", name = "Actual", marker= dict(size =3))])
fig.add_trace(go.Scatter3d(x= df['x1'], y = df['x2'], z = df['y_pred'], mode = "markers", name = "Actual", marker= dict(size =3)))
   # formate hai 3d image 
fig.show()
```

```python
poly = PolynomialFeatures(degree=3)
x_poly = poly.fit_transform(df[['x1','x2']])

p_model = LinearRegression()
p_model.fit(x_poly, df['y'])
```

```python
df['p_y_pred'] = p_model.predict(x_poly)

fig = go.Figure(data=[go.Scatter3d(x= df['x1'], y = df['x2'], z = df['y'], mode = "markers", name = "Actual", marker= dict(size =3))])
fig.add_trace(go.Scatter3d(x= df['x1'], y = df['x2'], z = df['p_y_pred'], mode = "markers", name = "Actual", marker= dict(size =3)))

fig.show()
```

 # for linear regession

```python
# y = mx + c ( loss km )
#MAE (mean_absolute_error)
from sklearn.metrics import mean_absolute_error  # mean milake ke liye 
print('MAE of LR:-',mean_absolute_error(df['y'],df['y_pred' ]))  # kisi bhi data frem ko get krne ke liye we are used to df[] this method

```

```python
# MSE (mean_squared_error)

from sklearn.metrics import mean_absolute_error  # mean milake ke liye
print('MSE of LR:-',mean_squared_error(df['y'],df['y_pred' ])) # same to same resion 
```

```python
#RMSE
print('RMSE of LR:-',np.sqrt(mean_squared_error(df['y'],df['y_pred' ])))
```

```python
#R2
from sklearn.metrics import r2_score
print('R2_score of LR:-',r2_score(df['y'],df['y_pred' ])) # same to same resion

```

for polynomial regression

```python
print('MAE of PR:-',mean_absolute_error(df['y'],df['p_y_pred' ])) 
print('MSE of PR:-',mean_squared_error(df['y'],df['p_y_pred' ]))
print('RMSE of PR:-',np.sqrt(mean_squared_error(df['y'],df['p_y_pred' ])))
print('R2_score of PR:-',r2_score(df['y'],df['p_y_pred' ]))
```
